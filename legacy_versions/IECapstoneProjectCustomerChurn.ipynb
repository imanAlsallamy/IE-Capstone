# How to read Box Plots:

Box plots (also called box-and-whisker plots) are a way to visually summarize the distribution of numerical data, highlighting key statistical features like quartiles, median, and potential outliers.

Here's a guide to understanding the different parts of a box plot:

1. The Box:

Median: The line inside the box represents the median (middle value) of the data. 50% of the data points are above this line, and 50% are below it.
Interquartile Range (IQR): The box itself spans the interquartile range, which is the difference between the third quartile (Q3) and the first quartile (Q1). The IQR contains the middle 50% of the data.
First Quartile (Q1): The bottom edge of the box represents the first quartile (Q1). 25% of the data points are below this value.
Third Quartile (Q3): The top edge of the box represents the third quartile (Q3). 75% of the data points are below this value.
2. The Whiskers:

Whiskers: The lines extending from the box are called whiskers. They represent the range of data within a certain distance from the quartiles (typically 1.5 times the IQR).
Upper Whisker: Extends to the highest data point within 1.5 * IQR from Q3.
Lower Whisker: Extends to the lowest data point within 1.5 * IQR from Q1.
3. Outliers:

Outliers: Data points that fall outside the whiskers are considered potential outliers. They are often plotted as individual points beyond the whiskers and can indicate unusual or extreme values in the dataset.
Interpreting a Box Plot:

By looking at a box plot, you can quickly get a sense of:

Central Tendency: The median gives you an idea of the typical or average value.
Spread and Variability: The IQR and the length of the whiskers show how spread out the data is. A shorter box and shorter whiskers indicate less variability, while a longer box and longer whiskers indicate more variability.
Skewness: If the median is closer to one end of the box than the other, or if one whisker is longer than the other, the data may be skewed. A longer right whisker suggests a right-skewed distribution (tail to the right), and a longer left whisker suggests a left-skewed distribution.
Outliers: The presence of outliers can indicate potential data errors, extreme values, or interesting observations that require further investigation.

```python
import pandas as pd
from scipy import stats

# Load your dataframe
file_path = '/content/drive/My Drive/IE University Capstone Project/churn.csv'
df = pd.read_csv(file_path)

# Select numerical columns for outlier detection
numerical_cols = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']

# Calculate Z-scores and identify outliers for each numerical column
for col in numerical_cols:
    df['z_scores_' + col] = stats.zscore(df[col])
    outliers = df[(df['z_scores_' + col] > 3) | (df['z_scores_' + col] < -3)]
    print(f"Outliers for {col}:")
    print(outliers[[col, 'z_scores_' + col]])  # Print outliers and their Z-scores
    print("\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats

# Load your dataframe
file_path = '/content/drive/My Drive/IE University Capstone Project/churn.csv'
df = pd.read_csv(file_path)

# Select numerical columns for distribution analysis
numerical_cols = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']

# Create histograms and Q-Q plots for each numerical column
for col in numerical_cols:
    # Histogram
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True)  # kde=True adds a kernel density estimate curve
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

    # Q-Q plot
    plt.figure(figsize=(8, 6))
    stats.probplot(df[col], dist="norm", plot=plt)  # Compare to normal distribution
    plt.title(f'Q-Q Plot of {col} vs. Normal Distribution')
    plt.show()
```
